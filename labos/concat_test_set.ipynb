```python
import os
import pandas as pd
```

```python
concatened_df = pd.DataFrame()
for genre in os.listdir('./extracted_test_set/'):
    for song_df in os.listdir(f'./extracted_test_set/{genre}/'):
        df = pd.read_csv(f'./extracted_test_set/{genre}/{song_df}')
        concatened_df = pd.concat([concatened_df, df], axis=0)
print(concatened_df)
concatened_df.to_csv('./resources/test_set.csv', index=False)
```
